Función para poner los 24 valores horarios por filas:

```python
def formatotabla_LSTM(calidad):
    global df0
    df0 = pd.DataFrame(columns=['ESTACION','CONTAMINANTE','AÑO','MES','DIA','HORA','VALOR'])
    df0['ESTACION'] = df0['ESTACION'].astype('str')
    df0['CONTAMINANTE'] = df0['CONTAMINANTE'].astype('str')
    df0['AÑO'] = df0['AÑO'].astype('int')
    df0['MES'] = df0['MES'].astype('int')
    df0['DIA'] = df0['DIA'].astype('int')
    df0['HORA'] = pd.to_numeric(df0['HORA'], errors='coerce').fillna(0)
    df0['VALOR'] = df0['VALOR'].astype('float64')
    df0.dtypes
    t=0
    for i in range(0,len(calidad)):
        estacion=calidad.iloc[i,0]
        contaminante=calidad.iloc[i,1]
        anho=calidad.iloc[i,4]
        mes=calidad.iloc[i,5]
        dia=calidad.iloc[i,6]
        hora=1
        for j in range(7,31):
            valor=calidad.iloc[i,j]
            linea = [str(estacion), str(contaminante), int(anho), int(mes), int(dia),int(hora),valor]
            df0.loc[t] = linea
            hora=hora+1
            t=t+1
    #print(df0.index.size)
    df0.head()
    #df0.to_csv('salida/stat_'+str(estacion)+'_cont'+str(contaminante)+'.csv', sep=',', encoding='utf-8', index=False)
```

```python
for estacion in estaciones_calidad_activas:
    for contaminante in contaminantes:
        calidad_filtrado=calidad[(calidad['MAGNITUD']==contaminante) & (calidad['ESTACION']==estacion)]
        formatotabla_LSTM(calidad_filtrado)
        tabla_LSTM=pd.merge(df0, trafico, how='inner', on=['AÑO','MES','DIA','HORA'])
        tabla_LSTM2=pd.merge(tabla_LSTM, aemetMadrid, how='inner', on=['AÑO','MES','DIA'])
        tabla_LSTM2.to_csv('salida/stat_'+str(estacion)+'_cont'+str(contaminante)+'.csv', sep=',', encoding='utf-8', index=False)
```
